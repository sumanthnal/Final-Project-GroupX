```python
import pandas as pd
import numpy as np
import nltk
import spacy
#import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data = pd.read_csv("mbti_1.csv")
# print(data.shape)
# print(data.info())
print(data['type'].value_counts())
```

```python
# Bar chart of personality type distribution
dist_table = data.groupby(['type']).count()

plt.figure(figsize=(8, 4))
plt.bar(np.array(dist_table.index), dist_table['posts'])
plt.xlabel('MBTI Types')
plt.ylabel('Count')
plt.title('Distribution of Personality Types')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(np.array(dist_table['posts']), 
        labels=np.array(dist_table.index), 
        autopct='%.1f%%',
        wedgeprops={'linewidth': 3.0, 
                    'edgecolor': 'white'})
plt.title("Pie Chart Comparison of MBTI Types")
plt.legend(title="MBTI Types")
```
